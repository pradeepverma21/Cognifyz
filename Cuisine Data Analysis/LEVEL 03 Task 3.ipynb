```python
import pandas as pd

# Load your dataset
data = pd.read_csv('cuisine_dataset.csv')


# Drop rows with missing values in relevant columns
data = data.dropna(subset=['Price range', 'Has Online delivery', 'Has Table booking'])

# Convert 'Price range' to numeric
data['Price range'] = pd.to_numeric(data['Price range'], errors='coerce')

# Create a pivot table to analyze the relationship
pivot_table = pd.pivot_table(data, values='Restaurant ID', index=['Price range'], columns=['Has Online delivery', 'Has Table booking'], aggfunc='count')

# Display the pivot table
print("Pivot Table:")
print(pivot_table)

# Alternatively, you can use groupby for a more concise representation
grouped_data = data.groupby(['Price range', 'Has Online delivery', 'Has Table booking']).size().reset_index(name='Count')

# Display the grouped data
print("\nGrouped Data:")
print(grouped_data)
```

```python
# Drop rows with missing values in relevant columns
data = data.dropna(subset=['Price range', 'Has Online delivery', 'Has Table booking'])

# Convert 'Price range' to numeric
data['Price range'] = pd.to_numeric(data['Price range'], errors='coerce')

# Convert 'Has Online delivery' and 'Has Table booking' to numeric (0 or 1)
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})
data['Has Table booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})

# Create a new DataFrame to calculate proportions
proportion_data = pd.DataFrame(columns=['Price range', 'Online delivery proportion', 'Table booking proportion'])
```

```python
# Calculate proportions for each price range
for price_range in data['Price range'].unique():
    subset = data[data['Price range'] == price_range]
    total_restaurants = len(subset)
    online_delivery_proportion = subset['Has Online delivery'].sum() / total_restaurants
    table_booking_proportion = subset['Has Table booking'].sum() / total_restaurants
    proportion_data = proportion_data.append({
        'Price range': price_range,
        'Online delivery proportion': online_delivery_proportion,
        'Table booking proportion': table_booking_proportion
    }, ignore_index=True)

# Plot the results
plt.figure(figsize=(12, 8))
sns.barplot(x='Price range', y='Online delivery proportion', data=proportion_data, color='blue', label='Online Delivery')
sns.barplot(x='Price range', y='Table booking proportion', data=proportion_data, color='green', label='Table Booking')
plt.title('Proportion of Restaurants Offering Online Delivery and Table Booking by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Proportion')
plt.legend()
plt.show()
```
